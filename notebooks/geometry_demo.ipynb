```julia
using Pkg; Pkg.activate("../tambo/")
using Plots
using StaticArrays
push!(LOAD_PATH, "../tambo/src/")
# using Position
```

```julia
prop_lwh = SVector{3}(3000, 3000, 4000)
prop_ct = CoordinateTransform([prop_lwh[i]/2 for i in eachindex(prop_lwh)]..., 0,0,0)
prop_box = Box(prop_lwh, prop_ct)

gen_lwh = SVector{3}(500, 1000, 300)
gen_ct = CoordinateTransform(3000, 3000, 3500, 0, 1, 0)
gen_box = Box(gen_lwh, gen_ct)
```

```julia
N = 10000
xx = (rand(N).-0.5).*gen_box.lwh[1]
yy = (rand(N).-0.5).*gen_box.lwh[2]
zz = (rand(N).-0.5).*gen_box.lwh[3]
p = Position.box_to_box.([SVector{3}(xyz) for xyz in zip(xx,yy,zz)], Ref(gen_box), Ref(prop_box));
```

```julia
# x view
plt = scatter3d([x[1] for x in p[:,1]],
          [x[2] for x in p[:,1]],
          [x[3] for x in p[:,1]],
          xlim=(0,3000),
          ylim=(0,3000),
          zlim=(0,4000),
)
plot!(plt, camera=(0, 0))

display(plt)

# Top view
plt = scatter3d([x[1] for x in p[:,1]],
          [x[2] for x in p[:,1]],
          [x[3] for x in p[:,1]],
          xlim=(0,3000),
          ylim=(0,3000),
          zlim=(0,4000),
)
plot!(plt, camera=(0, 90))
display(plt)

# y-view
plt = scatter3d([x[1] for x in p[:,1]],
          [x[2] for x in p[:,1]],
          [x[3] for x in p[:,1]],
          xlim=(0,3000),
          ylim=(0,3000),
          zlim=(0,4000),
)
plot!(plt, camera=(9, 0))
display(plt)
```
