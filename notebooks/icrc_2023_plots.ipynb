```julia
injection_events = simulator["injected_events"];
```

```julia
using Statistics
```

```julia
bins = LinRange(2, 9, 29)
```

```julia
zeniths .* 180 ./ π
```

```julia
θs = getfield.(injection_events["initial_state"]["direction"], :θ)
zeniths = acos.(LinRange(1, 0, 7))

    
colors = palette([:pink, :lightblue], 3)

kw = Dict(
        :fillrange=>[3000, 3000],
        :fillalpha=>0.03,
        :color=>:black
    )

plts = []

for (θ_min, θ_max) in zip(zeniths[1:end-1], zeniths[2:end])
    mask = θ_min .< θs .&& θs .< θ_max
    events = injection_events[mask]
    
    kw = Dict(
        :fillrange=>[3200, 3200],
        :fillalpha=>0.03,
        :color=>:black
    )

    
    plt = plot([2, 5], [0, 0]; kw...)
    # plt = plot([2, 5], [0, 0], fillrange=[300, 300], fillalpha=0.03, color=:black)
    plot!(plt, [7, 9], [0, 0]; kw...)
    
    label0 = ""
    label1 = ""
    label2 = ""
    if θ_min==0
        label0 = L"Injected $\nu_{\tau}$"
        label1 = L"TAMBO $\nu_{\tau}$"
        label2 = L"$\tau^{\pm}$"
    end
    

    stephist!(plt, log.(10, events["initial_state"]["energy"] / units.GeV), bins=bins, lw=2, c=colors[1], label=label0)
    stephist!(plt, log.(10, events["entry_state"]["energy"] / units.GeV), bins=bins, lw=2, c=colors[2], label=label1)
    stephist!(plt, log.(10, events["final_state"]["energy"] / units.GeV), bins=bins, lw=3, c=colors[3], label=label2)
    # xlims!(plt, 2, 9)
    # ylims!(plt, 0, 3200)
    # display(plt)
     
    title = (@sprintf "%.2f" cos(θ_min)) * L"$< \cos\,\theta <$" * (@sprintf "%.2f" cos(θ_max))
    title!(plt,  title)
    push!(plts, plt)

end

plt = plot(plts..., layout=(2,3), size=(900, 400))
```
